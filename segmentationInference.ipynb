```python
def dice_seg(input:Tensor, targs:Tensor, iou:bool=True, eps:float=1e-8)->Rank0Tensor:
    "Dice coefficient metric for binary target. If iou=True, returns iou metric, classic for segmentation problems."
    n = targs.shape[0]
    input = input.argmax(dim=1).view(n,-1)
    targs = targs.view(n,-1)
    intersect = (input * targs).sum().float()
    union = (input+targs).sum().float()
    if not iou: return (2. * intersect / union if union > 0 else union.new([1.]).squeeze())
    else: return (intersect / (union-intersect+eps) if union > 0 else union.new([1.]).squeeze())
```

```python
def segmentation_inference(learner, in_img_path, out_img_dir, show_image = True):
    img_fai = open_image(in_img_path)
    img = np.array((Image.open(in_img_path)))
    y = learn.predict(img_fai)
    y_np = y[1].numpy()
    y_np = y_np[0][:][:]
    y_out = np.stack((y_np,y_np,y_np),axis=2)
    img_out1 = (y_out.astype(np.float32))*(img.astype(np.float32))
    img_out = img_out1.astype(np.uint8)
    img_out[img_out == 0] = 255
    out = Image.fromarray(img_out, 'RGB')
    filename = in_img_path[-16:-1] + 'g'
    out.save(out_img_dir + '\\' + filename)
    if show_image: out.show()
```
